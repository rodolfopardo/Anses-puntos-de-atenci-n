```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

sns.set(rc={'figure.figsize':(20.7,8.27)})
```

```python
data = pd.read_csv('/Users/rodolfopardo/Downloads/anses.csv')
```

```python
data.head()
```

```python
data.Provincia.value_counts()
```

```python
sns.countplot(data.Provincia, order = data.Provincia.value_counts().head(10).index, palette = "Set2")
plt.title('Las 10 provincias con más sucursales de Anses', fontsize = 20)
plt.xlabel('Provincias', fontsize = 15)
plt.ylabel('Cantidad', fontsize = 15)
plt.show()
```

```python
mendoza = data[data['Provincia'] == "Mendoza"]
```

```python
mendoza['Punto de atención'].value_counts()
```

```python
sns.countplot(mendoza.Partido, order = mendoza.Partido.value_counts().index, palette = "Set2")
plt.title('Las 10 localidades de Mendoza con más sucursales', fontsize = 20)
plt.xlabel('Departamentos', fontsize = 15)
plt.ylabel('Cantidad', fontsize = 15)
plt.show()
```
